```python
AIM :
    The aim of data visualization in Python is to transform raw data into visual representations that facilitate insights and patterns identification. 
    It enables effective communication of data-driven findings to a wide range of audiences.
    
DESCRIPTION:
    Data visualization in Python involves representing data in visual formats such as graphs and charts to easily understand and interpret the information. 
    It utilizes libraries like Matplotlib, Seaborn, and Plotly to create various types of visualizations. The process includes preparing the data, selecting
    the appropriate visualization type, and customizing the visual elements to effectively convey insights.
```

```python
import matplotlib.pyplot as plt
import pandas as pd

loan_data = pd.read_csv('loan_prediction.csv')

# Bar chart with Property_area and its count
loan_data['Property_Area'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Property Area Count')
plt.xlabel('Property Area')
plt.ylabel('Count')
plt.show()

```

```python
# Comparative bar chart for “Graduate” and “Married” columns against Loan_status


loan_data.groupby(['Loan_Status', 'Married']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Loan Status based on Marital Status')
plt.show()
```

```python
# Horizontal bar chart for “Property_area” and “Loan_status”
loan_data.groupby(['Property_Area', 'Loan_Status']).size().unstack().plot(kind='barh', stacked=True)
plt.title('Loan Status based on Property Area')
plt.show()
```

```python
#  Stacked bar chart for Applicant_Income and Coapplicant Income against the Loan_status
loan_data.groupby('Loan_Status')[['ApplicantIncome', 'CoapplicantIncome']].mean().plot(kind='bar', stacked=True)
plt.title('Loan Status based on Income')
plt.show()
```

```python
#  Pie chart with Property_status and its count
loan_data['Property_Area'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Property Area Distribution')
plt.show()
```

```python
#  Dot plot between Gender and Loan_amount
plt.scatter(loan_data['Loan_ID'], loan_data['LoanAmount'], alpha=0.5)
plt.title('Dot Plot: Loan_ID vs Loan Amount')
plt.show()
```

```python
plt.plot(loan_data['ApplicantIncome'], loan_data['CoapplicantIncome'], marker='o', linestyle='-')
plt.title('Trend Line: ApplicantIncome vs CoapplicantIncome')
plt.show()
```

```python
#  Scatter plot between ApplicantIncome and Property_Area
plt.scatter(loan_data['ApplicantIncome'], loan_data['Property_Area'], alpha=0.5)
plt.title('Scatter Plot: ApplicantIncome vs Property Area')
plt.show()
```

```python
#  Scatter plot between Loan_Amount and Loan_Term
plt.scatter(loan_data['LoanAmount'], loan_data['Loan_Amount_Term'], alpha=0.5)
plt.title('Scatter Plot: Loan Amount vs Loan Term')
plt.show()
```

```python
#   Box plot for Loan_amount
plt.boxplot(loan_data['LoanAmount'].dropna())
plt.title('Box Plot: Loan Amount')
plt.show()
```

```python
#  Histogram plot for Property_Area column
plt.hist(loan_data['Property_Area'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram: Property Area')
plt.show()
```

```python
#  Histogram plot for Education column with bin value and PDF
plt.hist(loan_data['Education'], bins=2, color='skyblue', edgecolor='black', density=True)
plt.title('Histogram: Education')
plt.show()
```
